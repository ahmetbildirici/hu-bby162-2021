```python
import urllib.request
from bs4 import BeautifulSoup

kayitliAdresler = "/content/drive/MyDrive/kayitliadres.txt"
kayitliBilgiler = "/content/drive/MyDrive/kayitliBilgiler.txt"

def anaMenu():
  print("\033[43m║\t\033[43m\033[1mÖRÜMCEK'E HOŞ GELDİNİZ!\t\t\033[1m\033[43m║\033[43m")
  islem = input("\t1- Adres Girme\n\t2- Örümcek Çalıştırma\n\t3- .txt dosyasını temizleme\n\t4- Dosya yolunu ve içeriğini öğrenme\n\t5- Dosya boyutunu öğrenme\n\t6- Çıkış Yap\n   İşlem seçiniz: ")
  if islem == "1":
    adresGiris()
  elif islem == "2":
    orumcekCalistir()
  elif islem == "3":
    txtTemizle()
  elif islem == "4":
    dosyaYollari()
  elif islem == "5":
    dosyaBoyutlari()
  elif islem == "6":
    cikis()
  else:
    print("\033[41m\033[97mHatalı seçim yaptınız, tekrar yönlendiriliyorsunuz...")
    anaMenu()

def adresGiris():
    adreslerAc = open(kayitliAdresler, 'a')
    girilenAdresler = input("Adres giriniz: ")
    adreslerAc.write(girilenAdresler+"\n")
    adresTekrarSor()
    adreslerAc.close()
    adreslerAc = open(kayitliAdresler, 'r')
    for adresOku in adreslerAc:
      print(adresOku)
    adreslerAc.close()
    anaMenu()

def adresTekrarSor():
    adresSorusu = input("Başka adres girmek istiyor musunuz?\n-1- Evet\n-2- Hayır\nLütfen numara tuşlayınız: ")
    if adresSorusu == "1":
      adresGiris()
    elif adresSorusu == "2":
      print("\033[42m\033[97mAdresler başarıyla girildi.")
      anaMenu()
    else:
      print("\033[41m\033[97mHatalı işlem girdiniz. Sadece -1- ve -2- seçimini yapabilirsiniz.")
      adresTekrarSor()

def orumcekCalistir():
    adreslerAc = open(kayitliAdresler, 'r')
    bilgilerAc = open(kayitliBilgiler, 'w')
    for bilgiAl in adreslerAc:
      webSite = urllib.request.urlopen(bilgiAl)
      getBytes = webSite.read()
      webPage = getBytes.decode("utf8")
      webSite.close()
      soup = BeautifulSoup(webPage, 'html.parser')
      bilgilerAc.write(bilgiAl.strip() + " - " + soup.title.contents[0] + "\n")
    adreslerAc.close()
    bilgilerAc.close()
    bilgiOku = open(kayitliBilgiler, 'r')
    for bilgiGoster in bilgiOku:
      print(bilgiGoster)
      bilgiOku.close()
      anaMenu()

def cikis():
  print("1) Evet\n2) Hayır\nÇıkış yapmak istiyor musunuz? ")
  cikisIslem = input("\nİşlem seçiniz: ")
  if cikisIslem == "2":
    print("Ana menüye yönlendiriliyorsunuz...")
    anaMenu()
  elif cikisIslem == "1":
    print("\033[90mÇıkış yaptınız. Teşekkürler.\033[90m")
  else:
    print("\033[41m\033[97mHatalı işlem girdiniz. Sadece -1- ve -2- seçimini yapabilirsiniz.")
    cikis()

def txtTemizle():
  temizlemeHedefi = input("\t1) Adresler\n\t2) Bilgiler\nTemizlenecek dosyayı seçiniz: ")
  if temizlemeHedefi == "1":
    open(kayitliAdresler, "w").close()
    print("\033[42m\033[97mDosya başarıyla temizlendi, ana menüye yönlendiriliyorsunuz...")
    anaMenu()
  elif temizlemeHedefi == "2":
    open(kayitliBilgiler, "w").close()
    print("\033[42m\033[97mDosya başarıyla temizlendi, ana menüye yönlendiriliyorsunuz...")
    anaMenu()
  else:
    print("\033[41m\033[97mHatalı işlem girdiniz... Fonksiyona tekrar yönlendiriliyorsunuz.")
    txtTemizle()

def dosyaBoyutlari():
  boyutOgren = input("1) Adresler\n2) Bilgiler\nBoyutunu öğrenmek istediğiniz dosyanın numarasını giriniz:")
  if boyutOgren == "1":
    import os
    b = os.path.getsize(kayitliAdresler)
    print(b)
    print("BYTE")
    anaMenu()
  elif boyutOgren == "2":
    import os
    b = os.path.getsize(kayitliBilgiler)
    print(b)
    print("BYTE")
    anaMenu()
  else:
    print("\033[41m\033[97mHatalı işlem girdiniz... Fonksiyona tekrar yönlendiriliyorsunuz.")
    dosyaBoyutlari()

def dosyaYollari():
  yolOgren = input("\t1) Adresler\n\t2) Bilgiler\nYolunu ve içeriğini öğrenmek istediğiniz dosyanın numarasını giriniz: ")
  if yolOgren == "1":
    with open(kayitliAdresler) as dosya:
          print(dosya.read())
    print("Dosya Yolu: ",dosya.name)
    anaMenu()
  elif yolOgren == "2":
    with open(kayitliBilgiler) as dosya:
          print(dosya.read())
    print("Dosya Yolu: ",dosya.name)
    anaMenu()
  else:
    print("\033[41m\033[97mHatalı işlem girdiniz... Fonksiyona tekrar yönlendiriliyorsunuz.")
    dosyaYollari()

anaMenu()
```
